# 1. Perceptron Mistakes

```python
import numpy as np
```

```python
def perceptron_mistakes(points):

    #the first point that the algorithm sees is always considered a mistake

    count_mistakes=0
    progression=[]
    theta=np.array([0,0])
    
    for t in range(10):
        for i in points:
            theta_old=theta
            if i[1]*(np.dot(theta,i[0]))<=0:
                theta=theta+i[1]*i[0]
                count_mistakes+=1
                progression.append(list(theta))
            
            print("Theta old:",theta_old," | x:",i[0]," | y:",i[1]," | Condition:",i[1]*(np.dot(theta_old,i[0]))," | Theta new:",theta)
            
    print("Number of mistakes:",count_mistakes)
    print("Progression of theta:",progression)
```

## 1a) Start with x(1)

```python
x1=[np.array([-1,-1]),1]
x2=[np.array([1,0]),-1]
x3=[np.array([-1,1.5]),1]

perceptron_mistakes([x1,x2,x3])
```

## 6. Working out Perceptron Algorithm

```python
def perceptron_mistakes_6(points):

    #the first point that the algorithm sees is always considered a mistake

    count_mistakes=0
    progression=[]
    theta=np.array([0,0,0])
    
    for t in range(10):
        for i in points:
            theta_old=theta
            if np.dot(theta,i[0])==0:
                theta=theta+i[1]*i[0]
                count_mistakes+=1
                progression.append(list(theta))
            
            print("Theta old:",theta_old," | x:",i[0]," | y:",i[1]," | Condition:",i[1]*(np.dot(theta_old,i[0]))," | Theta new:",theta)
            
    print("Number of mistakes:",count_mistakes)
    print("Progression of theta:",progression)
```
